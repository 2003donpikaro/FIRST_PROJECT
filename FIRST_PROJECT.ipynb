```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data =pd.read_csv(r"C:\Users\R.M\Desktop\az system\cell_samples.csv")
data.head()
data = data.dropna()
for i in data.columns:
    data[i]=pd.to_numeric(data[i],errors='coerce').fillna(0)
```

```python
data['Class'].value_counts()
```

```python
x = np.asanyarray(data.iloc[:,:-1])
y = np.asanyarray(data.iloc[:,-1])
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2,random_state=42)
print(xtrain.shape,ytrain.shape)
print(xtest.shape,ytest.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,f1_score
neighbor =KNeighborsClassifier(n_neighbors=5)
neighbor.fit(xtrain,ytrain)
yhat = neighbor.predict(xtest)
print(accuracy_score(ytest,yhat))
print(precision_score(ytest,yhat,pos_label=4))
print(recall_score(ytest,yhat,pos_label=4))
print(confusion_matrix(ytest,yhat))
print(f1_score(ytest,yhat,pos_label=4))
```

```python
import seaborn as sns
cm=confusion_matrix(ytest,yhat)
sns.heatmap(cm,annot =True,fmt ='d',cmap='coolwarm' )
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
model.fit(xtrain,ytrain)
yhat = model.predict(xtest)
print(accuracy_score(ytest,yhat))
print(recall_score(ytest,yhat,pos_label=4))
print(precision_score(ytest,yhat,pos_label =4))
```

```python
cm =confusion_matrix(ytest,yhat)
sns.heatmap(cm,annot=True,fmt ='d',cmap='viridis',linecolor='green')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
scaler =DecisionTreeClassifier(max_depth=5)
scaler.fit(xtrain,ytrain)
yhat = scaler.predict(xtest)
print(accuracy_score(ytest,yhat))
print(precision_score(ytest,yhat,pos_label=4))
print(recall_score(ytest,yhat,pos_label=4))
```

```python
cm =confusion_matrix(ytest,yhat)
sns.heatmap(cm,annot=True,fmt ='.2g',cmap='Blues')
plt.plot()
```

```python
from sklearn.svm import SVC
scaler =SVC(kernel ='rbf')
scaler.fit(xtrain,ytrain)
yaht = scaler.predict(xtest)
print(accuracy_score(ytest,yhat))
print(recall_score(ytest,yaht,pos_label=4))
print(precision_score(ytest,yaht,pos_label=4))
```
